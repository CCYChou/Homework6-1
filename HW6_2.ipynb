```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
import numpy as np
import requests
from PIL import Image
import io
import os
import matplotlib.pyplot as plt
import wget

def download_dataset():
    base_url = "https://raw.githubusercontent.com/prajnasb/observations/master/experiements/data"
    classes = ['with_mask', 'without_mask']

    if not os.path.exists('data'):
        os.makedirs('data')
        for class_name in classes:
            class_path = f'data/{class_name}'
            if not os.path.exists(class_path):
                os.makedirs(class_path)

            response = requests.get(f"https://api.github.com/repos/prajnasb/observations/contents/experiements/data/{class_name}")
            files = [item['download_url'] for item in response.json() if item['name'].endswith('.jpg')]

            print(f"Downloading {class_name} images...")
            for i, file_url in enumerate(files):
                wget.download(file_url, f'{class_path}/image_{i}.jpg')
            print()

def analyze_dataset():
    for class_name in ['with_mask', 'without_mask']:
        path = f'data/{class_name}'
        num_images = len([f for f in os.listdir(path) if f.endswith('.jpg')])
        print(f"{class_name}: {num_images} images")

def prepare_data():
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        validation_split=0.2
    )

    train_generator = train_datagen.flow_from_directory(
        'data',
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary',
        subset='training'
    )

    validation_generator = train_datagen.flow_from_directory(
        'data',
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary',
        subset='validation'
    )

    return train_generator, validation_generator

def build_model():
    base_model = VGG16(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )

    for layer in base_model.layers:
        layer.trainable = False

    x = Flatten()(base_model.output)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=base_model.input, outputs=predictions)
    return model

def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(history.history['accuracy'], label='Training')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()

    ax2.plot(history.history['loss'], label='Training')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()

    plt.show()

def evaluate_model(model, validation_generator):
    evaluation = model.evaluate(validation_generator)
    print(f"\nValidation Loss: {evaluation[0]:.4f}")
    print(f"Validation Accuracy: {evaluation[1]:.4f}")
    return evaluation

def test_image(image_url, model):
    try:
        response = requests.get(image_url)
        img = Image.open(io.BytesIO(response.content))
        img = img.resize((224, 224))
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0

        prediction = model.predict(img_array)
        predicted_class = "With Mask" if prediction[0][0] > 0.5 else "Without Mask"
        confidence = prediction[0][0] if prediction[0][0] > 0.5 else 1 - prediction[0][0]

        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Prediction: {predicted_class}\nConfidence: {confidence:.2%}")
        plt.show()

        print(f"Predicted class: {predicted_class}")
        print(f"Confidence: {confidence:.2%}")

    except Exception as e:
        print(f"Error processing image: {str(e)}")

def save_model(model, path='mask_detection_model.h5'):
    model.save(path)
    print(f"Model saved to {path}")

def main():
    print("Downloading dataset...")
    download_dataset()

    print("\nAnalyzing dataset...")
    analyze_dataset()

    print("\nPreparing data...")
    train_generator, validation_generator = prepare_data()

    print("\nBuilding model...")
    model = build_model()
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    model.summary()

    print("\nTraining model...")
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size,
        epochs=30
    )

    plot_training_history(history)
    evaluate_model(model, validation_generator)
    save_model(model)

    print("\nModel ready for testing!")
    while True:
        image_url = input("\nEnter image URL (or 'quit' to exit): ")
        if image_url.lower() == 'quit':
            break
        test_image(image_url, model)

if __name__ == "__main__":
    main()
```

```python
import gradio as gr
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import torch

class AIApplication:
    def __init__(self):
        # Initialize models
        self.sentiment_analyzer = pipeline("sentiment-analysis")
        self.summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
        self.translator = pipeline("translation", model="Helsinki-NLP/opus-mt-zh-en")
        self.qa_model = pipeline("question-answering", model="deepset/roberta-base-squad2")

    def analyze_sentiment(self, text):
        result = self.sentiment_analyzer(text)[0]
        return f"情感: {result['label']}, 信心度: {result['score']:.2%}"

    def generate_summary(self, text):
        summary = self.summarizer(text, max_length=130, min_length=30)[0]
        return summary['summary_text']

    def translate_text(self, text):
        translation = self.translator(text)[0]
        return translation['translation_text']

    def answer_question(self, context, question):
        result = self.qa_model(question=question, context=context)
        return f"答案: {result['answer']}\n信心度: {result['score']:.2%}"

def create_interface():
    app = AIApplication()

    # Create interface
    with gr.Blocks(title="AI 多功能應用") as interface:
        gr.Markdown("# AI 多功能應用")

        with gr.Tab("情感分析"):
            with gr.Column():
                sentiment_input = gr.Textbox(label="輸入文字")
                sentiment_button = gr.Button("分析")
                sentiment_output = gr.Textbox(label="分析結果")
                sentiment_button.click(app.analyze_sentiment,
                                    inputs=sentiment_input,
                                    outputs=sentiment_output)

        with gr.Tab("文本摘要"):
            with gr.Column():
                summary_input = gr.Textbox(label="輸入長文本", lines=5)
                summary_button = gr.Button("生成摘要")
                summary_output = gr.Textbox(label="摘要結果")
                summary_button.click(app.generate_summary,
                                  inputs=summary_input,
                                  outputs=summary_output)

        with gr.Tab("中英翻譯"):
            with gr.Column():
                translate_input = gr.Textbox(label="輸入中文")
                translate_button = gr.Button("翻譯")
                translate_output = gr.Textbox(label="英文翻譯")
                translate_button.click(app.translate_text,
                                    inputs=translate_input,
                                    outputs=translate_output)

        with gr.Tab("問答系統"):
            with gr.Column():
                qa_context = gr.Textbox(label="背景內容", lines=5)
                qa_question = gr.Textbox(label="問題")
                qa_button = gr.Button("回答")
                qa_output = gr.Textbox(label="答案")
                qa_button.click(app.answer_question,
                              inputs=[qa_context, qa_question],
                              outputs=qa_output)

    return interface

def main():
    interface = create_interface()
    interface.launch()

if __name__ == "__main__":
    main()
```
